# **[imputation for missing values](https://)**

```python
# Impute missing values for numeric columns using media
for column in numeric_columns:
    df[column] = df[column].fillna(df[column].median())

# Impute missing values for non-numeric columns using mode
for column in categorical_columns:
    df[column] = df[column].fillna(df[column].mode()[0])
```

# **[Handling outliers](https://)**

```python
#Detecting Outliers by Using Interquartile Range and Boxplots
import matplotlib.pyplot as plt
import seaborn as sns

columns_to_check = ["Waist Circumference", "Pulmonary Function"]
for col in columns_to_check:
    fig, axes = plt.subplots(figsize=(8, 6))

    sns.boxplot(data=df, x=col, ax=axes) # Plot boxplot before handling outliers
    axes.set_title(f'Boxplot of {col} before outlier removal')

    plt.tight_layout()
```

```python
# Compute percentiles using Pandas quantile() function
percentile_25 = df['Waist Circumference'].quantile(0.25)
percentile_50 = df['Waist Circumference'].quantile(0.5)
percentile_75 = df['Waist Circumference'].quantile(0.75)

print("25th Percentile:", percentile_25)
print("50th Percentile:", percentile_50)
print("75th Percentile:", percentile_75)

# Compute interquartile range (IQR)
iqr = percentile_75 - percentile_25
print("Interquartile Range (IQR):", iqr)

# Calculate lower and upper bounds
lower_bound_Waist_Circumference = percentile_25 - 1.5 * iqr
upper_bound_Waist_Circumference = percentile_75 + 1.5 * iqr
print(f"Lower Bound for Waist Circumference = {lower_bound_Waist_Circumference}, and Upper Bound for fare = {upper_bound_Waist_Circumference}")

# Repeat the process for the Pulmonary_Function' column
percentile_25_Pulmonary_Function = df['Pulmonary Function'].quantile(0.25)
percentile_50_Pulmonary_Function = df['Pulmonary Function'].quantile(0.5)
percentile_75_Pulmonary_Function = df['Pulmonary Function'].quantile(0.75)

print("25th Percentile for Pulmonary Function:", percentile_25_Pulmonary_Function)
print("50th Percentile for Pulmonary Function:", percentile_50_Pulmonary_Function)
print("75th Percentile for Pulmonary Function:", percentile_75_Pulmonary_Function)

# Compute interquartile range (IQR) for Pulmonary_Function
iqr_Pulmonary_Function = percentile_75_Pulmonary_Function - percentile_25_Pulmonary_Function
print("Interquartile Range (IQR) for Pulmonary Function:", iqr_Pulmonary_Function)

# Calculate lower and upper bounds for Pulmonary_Function
lower_bound_Pulmonary_Function = percentile_25_Pulmonary_Function - 1.5 * iqr_Pulmonary_Function
upper_bound_Pulmonary_Function = percentile_75_Pulmonary_Function + 1.5 * iqr_Pulmonary_Function
print(f"Lower Bound for Pulmonary Function = {lower_bound_Pulmonary_Function}, and Upper Bound for Pulmonary Function = {upper_bound_Pulmonary_Function}")
```

```python
# Calculate the number of records before outlier removal
num_records_before = df.shape[0]

# Calculate the number of outliers based on the Interquartile Range (IQR) for the 'Pulmonary_Function' column
lower_bound_Pulmonary_Function = percentile_25_Pulmonary_Function - 1.5 * iqr_Pulmonary_Function
upper_bound_Pulmonary_Function = percentile_75_Pulmonary_Function + 1.5 * iqr_Pulmonary_Function
dropped_records_Pulmonary_Function = df[(df['Pulmonary Function'] < lower_bound_Pulmonary_Function) | (df['Pulmonary Function'] > upper_bound_Pulmonary_Function)].shape[0]
print(f"Number of outliers based on the Interquartile Range for 'Pulmonary_Function' column: {dropped_records_Pulmonary_Function} ({100 * dropped_records_Pulmonary_Function / num_records_before:.2f}%)")

# Repeat the process for the 'fare' column
lower_bound_Waist_Circumference = percentile_25 - 1.5 * iqr
upper_bound_Waist_Circumference = percentile_75 + 1.5 * iqr
dropped_records_Waist_Circumference = df[(df['Waist Circumference'] < lower_bound_Waist_Circumference) | (df['Waist Circumference'] > upper_bound_Waist_Circumference)].shape[0]
print(f"Number of outliers based on the Interquartile Range for 'Waist_Circumference' column: {dropped_records_Waist_Circumference} ({100 * dropped_records_Waist_Circumference / num_records_before:.2f}%)")
```

```python
# Apply outlier removal based on the calculated lower and upper bounds for the 'age' column
df = df[(df['Pulmonary Function'] >= lower_bound_Pulmonary_Function) & (df['Pulmonary Function'] <= upper_bound_Pulmonary_Function)]

# Apply outlier removal based on the calculated lower and upper bounds for the 'fare' column
df = df[(df['Waist Circumference'] >= lower_bound_Waist_Circumference) & (df['Waist Circumference'] <= upper_bound_Waist_Circumference)]
```

```python
import time
import tracemalloc  # Import tracemalloc for memory tracking
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

target_column = 'Target'

# Split the data into features (principal components) and target
X_pca = df_pca.drop(columns=[target_column])
y_pca = df_pca[target_column]

# Split into training and testing subsets
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y_pca, test_size=0.2, random_state=42)

# --- Support Vector Machine (SVM) ---
tracemalloc.start()  # Start tracking memory usage
start_time = time.time()
svm_model = SVC(kernel='rbf', random_state=42)  # Specify 'rbf' kernel
svm_model.fit(X_train_pca, y_train_pca)  # Use the training subset
end_time = time.time()
current_memory, peak_memory = tracemalloc.get_traced_memory()  # Get memory usage
tracemalloc.stop()  # Stop memory tracking

# Calculate training time
svm_training_time = end_time - start_time

# Predict using the SVM model on the test data
y_pred_svm = svm_model.predict(X_test_pca)

# Evaluate the SVM model
svm_accuracy = accuracy_score(y_test_pca, y_pred_svm)
svm_precision = precision_score(y_test_pca, y_pred_svm, average='weighted')
svm_recall = recall_score(y_test_pca, y_pred_svm, average='weighted')
svm_f1 = f1_score(y_test_pca, y_pred_svm, average='weighted')

# Print the results for SVM
print("\nSupport Vector Machine (SVM):")
print(f"  Training Time: {svm_training_time:.4f} seconds")
print(f"  Peak Memory Usage: {peak_memory / 10**6:.4f} MB")
print(f"  Accuracy: {svm_accuracy:.4f}")
print(f"  Precision: {svm_precision:.4f}")
print(f"  Recall: {svm_recall:.4f}")
print(f"  F1-Score: {svm_f1:.4f}")

# Confusion Matrix Visualization
cm_pca = confusion_matrix(y_test_pca, y_pred_svm)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_pca, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix SVM (after PCA)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
